```python
import matplotlib.pyplot as p
import pandas as pd

df=pd.read_csv(r"C:\Users\venuk\Documents\indiapop.csv")
x=df['Age Group']
y=df['Population (Millions)']

p.figure(figsize=(12,6))
b=p.bar(x,y,color='steelblue')
p.xlabel('Age Group')
p.ylabel('Population (in Millions)')
p.title("India's Age-Wise Polulation Distribution (2022)")
p.xticks(rotation=45)
p.grid(axis='y',linestyle='--',alpha=0.6)
for bar in b:
    yval = bar.get_height()
    p.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval:.1f}M', ha='center', va='bottom')

p.tight_layout()
p.show()
```
